```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#display all the columns

pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv(r'dataset/train.csv')
```

```python
dataset.head(5)
```

```python
#capture all nan values of categorical features

cat_cols_nan = [i for i in dataset.columns if dataset[i].isnull().sum() > 1 and dataset[i].dtypes == 'O']

for i in cat_cols_nan:
    print(i, np.round(dataset[i].isnull().mean(), 4), '% missing values')
```

```python
#replace the categorical features nan values with missing

def replace_nan(dataset, cat_cols_nan):
    data = dataset.copy()
    data[cat_cols_nan] = data[cat_cols_nan].fillna('Missing')
    return data

dataset = replace_nan(dataset, cat_cols_nan)
dataset[cat_cols_nan].isnull().sum()
```

```python
dataset.head()
```

```python
numerical_nan = [i for i in dataset.columns if dataset[i].isnull().sum() > 1 and i not in cat_cols_nan]

for i in numerical_nan:
    print(i, np.round(dataset[i].isnull().mean(), 4), '% missing values')
```

```python
#replace the numerical value nan values with median cz there are some outliers

for i in numerical_nan:
    median_val = dataset[i].median()
    print(median_val)
    
    #create a new col of all nan values
    dataset[i+'nan'] = np.where(dataset[i].isnull(), 1, 0)
    dataset[i].fillna(median_val, inplace=True)
    
    
dataset[numerical_nan].isnull().sum()
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
#Datetime columns

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
    
```

```python
dataset.head()
```

```python
dataset.select_dtypes(exclude='object').skew()
```

```python
# since we have skewed numerical cols we need to perform log distribution cz skewed means it is not in guassian distribution form

num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice'] # these features don't have 0 value

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

```python
cat_features = [i for i in dataset.columns if dataset[i].dtypes == 'O']
```

```python
cat_features
```

```python
for feature in cat_features:
    temp = dataset.groupby(feature)['SalePrice'].count() / len(dataset)
    temp_df = temp[temp > 0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'Rare_var')
```

```python
dataset.head()
```

```python
# [dataset[feature].astype('category') for feature in categorical_features]
for feature in cat_features:
    dataset[feature] = dataset[feature].astype('category')
```

```python
cat_columns = dataset.select_dtypes(['category']).columns
```

```python
cat_columns
```

```python
dataset[cat_columns] = dataset[cat_columns].apply(lambda x: x.cat.codes)
```

```python
dataset.head()
```

```python
# Feature Scaling

feature_scale = [feature for feature in dataset.columns if feature not in ['Id', 'SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
# transform the train and test set and add on the Id and SalePrice variable
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True), 
                     pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)], 
                     axis=1)
```
